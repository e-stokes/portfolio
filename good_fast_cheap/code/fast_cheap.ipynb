```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## producing a histogram of all our numeric variables
df.hist(figsize=(18, 9), layout=(2, 3));
```

```python
## creating a binarized 'wage'
df['wage'] = pd.get_dummies(df['wage'], drop_first = True)
```

```python
## creating our numerical features mask
numerical_features = ['age', 'capital-gain', 'capital-loss', 'education-num', 'fnlwgt', 'hours-per-week']
```

```python
## creating boxen plots of our numerical features
for feature in numerical_features:
    sns.catplot(x='wage', y=feature, kind="boxen", data=df);
```

```python
## creating categorical plot of our numerical features
for feature in numerical_features:

    sns.catplot(x="wage", y=feature, data=df);
```

```python
## creating a mask of our categorical features
cat_columns = df.select_dtypes(include=['object']).columns
```

```python
## creating distribution plots of our categorical values

cat_plot = sns.FacetGrid(pd.melt(df, value_vars=sorted(cat_columns)), aspect=2, height=8,
                         size=10, col='variable', col_wrap=2, sharex=False, sharey=False)

cat_plot = cat_plot.map(sns.countplot,'value', )

## list comprehension to label the x-axes
[plt.setp(ax.get_xticklabels(), rotation=45) for ax in cat_plot.axes.flat]

## using 
cat_plot.fig.tight_layout()
plt.show()
```

```python
## showing correlation
plt.figure(figsize=(8,12))
sns.heatmap(df.corr()[['wage']].sort_values('wage'), vmin=-1, annot=True, cmap='coolwarm');
```
